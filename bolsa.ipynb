```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'evolucion-precios-acciones.xls')
```

### Banco de Guayaquil

```python
bg = df.iloc[11]
```

```python
bg_valores = np.array(bg[1:])
bg_tiempo = np.arange(len(bg_valores))
```

```python
plt.plot(bg_tiempo, bg_valores, color='purple')
```

### Banco Pichincha

```python
bp = df.iloc[12]
```

```python
bp = bp.replace('NMP', method='ffill')
```

```python
bp_valores = np.array(bp[1:])
```

```python
plt.plot(bg_tiempo, bp_valores, color='yellow')
```

### Banco Guayaquil vs Banco Pichincha

```python
bancos_a = np.array([bg_valores, bp_valores])
bancos_a_o = [[bancos_a[0][i], bancos_a[1][i]] for i in range(25)]
```

```python
bancos = pd.DataFrame(bancos_a_o, columns=[bg[0], bp[0]])
```

```python
fig, ax = plt.subplots()
ax.plot(bg_tiempo, bancos['Banco Guayaquil']*100, color='purple', label='Banco Guayaquil')
ax.plot(bg_tiempo, bancos['Banco Pichincha'], color='gold', label='Banco Pichincha')
plt.legend()
```
